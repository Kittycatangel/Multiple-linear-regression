# Import libaries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Load data

```python
from sklearn.datasets import load_boston
dataset = load_boston()
```

# Data exploration

```python
print(dataset.data)
```

```python
print(dataset.feature_names)
```

```python
print(dataset.DESCR)
```

```python
print(dataset.target)
```

# Transform data into Pandas Dataframe

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
df['PRICE'] = dataset.target
df.head()
```

# Data cleaning

```python
df.info()
```

```python
print(df.isnull().sum())
```

# Feature selection

```python
corr = df.corr()
print(corr)
```

```python
# get the top 3 features that has the highest correlation
print(df.corr().abs().nlargest(3, 'PRICE').index)
```

```python
# print the top 3 correlation values
print(df.corr().abs().nlargest(3, 'PRICE').values[:,13])
```

# Multiple regression

```python
plt.scatter(df['LSTAT'], df['PRICE'], marker='o')
plt.xlabel('LSTAT')
plt.ylabel('PRICE')
```

```python
plt.scatter(df['RM'], df['PRICE'], marker='o')
plt.xlabel('RM')
plt.ylabel('PRICE')
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(18,15))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['LSTAT'],
df['RM'],
df['PRICE'],
c='b')
ax.set_xlabel("LSTAT")
ax.set_ylabel("RM")
ax.set_zlabel("PRICE")
plt.show()
```

# Model Training

```python
x = pd.DataFrame(np.c_[df['LSTAT'], df['RM']], columns = ['LSTAT','RM'])
Y = df['PRICE']
```

## Split data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, Y_train, Y_test = train_test_split(x, Y, test_size = 0.3,
random_state=5)
```

```python
print(x_train.shape)
print(Y_train.shape)
```

```python
print(x_test.shape)
print(Y_test.shape)
```

## Model training

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, Y_train)
```

## Model prediction

```python
price_pred = model.predict(x_test)
```

```python
print('R-Squared: %.4f' % model.score(x_test,
Y_test))
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_test, price_pred)
print(mse)
plt.scatter(Y_test, price_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted prices")
plt.title("Actual prices vs Predicted prices")
```

## Intercept & coefficients

```python
print(model.intercept_)
print(model.coef_)
```

## Make a prediction

```python
#We can use the model to make a prediction for the house price when LSTAT
#is 30 and RM is 5

print(model.predict([[30,5]]))
```
